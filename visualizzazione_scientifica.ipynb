```python
import pandas as pd
import IPython.display
import numpy as np
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive/')
```

```python
data_dir = '/content/drive/MyDrive/datavis/'
```

```python
csv = pd.read_csv(data_dir + 'women_in_parliament.csv')
csv = csv.drop('Last Election Date', axis=1)
csv = csv.drop('Last Election Date footnote', axis=1)
csv = csv.drop('Series', axis=1)
csv = csv.drop('Footnotes', axis=1)
csv = csv.drop('Source', axis=1)

italy = csv[csv["Region"] == "Italy"]
germany = csv[csv["Region"] == "Germany"]
france = csv[csv["Region"] == "France"]
spain = csv[csv["Region"] == "Spain"]

neurope = csv[csv["Region"] == "Northern Europe"]
seurope = csv[csv["Region"] == "Southern Europe"]
weurope = csv[csv["Region"] == "Western Europe"]
eeurope = csv[csv["Region"] == "Eastern Europe"]
europe = [neurope, seurope, weurope, eeurope]
europe = pd.concat(europe)


italy = italy.to_numpy()
germany = germany.to_numpy()
france = france.to_numpy()
spain = spain.to_numpy()

europe = pd.pivot(europe, index='Region', columns='Year', values='Value')
display(europe)
```

```python
plt.subplot(121)
plt.plot(italy[:,1], italy[:, 2], label=italy[0, 0])
plt.plot(germany[:,1], germany[:, 2], label=germany[0, 0])
plt.plot(france[:,1], france[:, 2], label=france[0, 0])
plt.plot(spain[:,1], spain[:, 2], label=spain[0, 0])
plt.legend()

plt.subplot(122)
for i in range(4):
  plt.plot(europe[])

plt.show()
```
